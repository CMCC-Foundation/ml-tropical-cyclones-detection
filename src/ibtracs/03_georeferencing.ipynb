```python
import numpy as np
import xarray as xr
import pandas as pd

import sys
sys.path.append('../../resources/library/')
import tropical_cyclone as tc
```

```python
ds = xr.open_mfdataset('../../data/template.nc')
ibtracs = pd.read_csv('../../data/ibtracs/filtered/ibtracs_main-tracks_6h_1980-2021_TS-NR-ET-MX-SS-DS.csv', index_col=0)
```

```python
# get latitudes and longitudes
lats, lons = ds['lat'].data, ds['lon'].data
```

```python
georef = ibtracs.copy()
georef = georef[['ISO_TIME','SID','LAT','LON']]
georef['RLAT'] = tc.georef.round_to_grid(georef['LAT'])
georef['RLON'] = tc.georef.round_to_grid(georef['LON'])
georef['YLAT'] = [np.argwhere(lats==l)[0][0] for l in georef['RLAT']]
georef['XLON'] = [np.argwhere(lons==l)[0][0] for l in georef['RLON']]
georef
```

```python
georef.to_csv('../../data/ibtracs/georef.csv')
```
